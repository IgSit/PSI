```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
df = df.groupby(["Monitor", "ClientIterations", "SchedulerIterations"]).agg(["mean", "std"])


df
```

```python
df_lc = df.loc['ThreeLock'].reset_index()
df_ao = df.loc['ActiveObject'].reset_index()

df_lc
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import LinearLocator
import numpy as np
```

```python
def draw_plot(df_lc, 
              df_ao,
              lclabel: str,
              aolabel: str,
              title: str,
              xlabel: str,
              ylabel: str,
              zlabel: str,
              value: str,
              view1: int = 70,
              view2: int = 70):
    # LOCKS

    fig = plt.figure(figsize = (8, 8))
    ax = fig.add_subplot(111,
        projection='3d', 
        title=title,
        xlabel=xlabel,
        ylabel=ylabel,
        zlabel=zlabel
    )
    ax.view_init(view1, view2)
    ax.zaxis.set_major_locator(LinearLocator(5))

    # Make data.
    df = df_lc
    X = df["SchedulerIterations"].unique()
    xlen = len(X)
    Y = df["ClientIterations"].unique()
    ylen = len(Y)
    Z = [
            [
            (list(df.loc[(df['SchedulerIterations'] == i) & (df['ClientIterations'] == j)][value]["mean"].values) or [0])[0]
                for i in X
        ] for j in Y 
    ]
    Z = np.array(Z)
    X,Y = np.meshgrid(X, Y)

    # Plot data
    locks = ax.plot_surface(X, Y, Z, color="red", label="locks")
    
    # AO

    # Make data.
    df = df_ao
    X = np.array(df["SchedulerIterations"].unique())
    Y = np.array(df["ClientIterations"].unique())
    Z = [
            [
            (list(df.loc[(df['SchedulerIterations'] == i) & (df['ClientIterations'] == j)][value]["mean"].values) or [0])[0]
                for i in X
        ] for j in Y 
    ]
    Z = np.array(Z)
    X,Y = np.meshgrid(X, Y)

    # Plot data
    ao = ax.plot_surface(X, Y, Z, color="blue", label="ao")
        
    plt.show()
```

```python
draw_plot(df_lc, 
          df_ao,
          "Three Lock",
          "Active Object",
          "Total Change",
          "Scheduler computations",
          "Client Computations",
          "Total Change",
          "TotalChange",
         10,110)
```
